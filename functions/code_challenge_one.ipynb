```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = np.arange(-3.0, 3.0, 0.1)
f = [x**2 + 3*x**3 -x**4 for x in X]
```

```python
plt.plot(X, f )
plt.grid(True)
plt.ylabel('y')
plt.xlabel('x')
plt.title('x**2 + 3x**3-x**4')
plt.show()
```
