```python
from sympy import *
import sympy as sym
```

```python
b = symbols('b')
f = -b**4+3*b**3+b**2
```

```python
f
```

```python
from sympy.plotting import plot

p1 = plot(f, label="$-\beta ^4+3\beta ^3+\beta ^2$", legend=True, size=(23,23))
```

Teacher solution

```python
s_beta = sym.var('beta')
s_y = -s_beta**4+3*s_beta**3+s_beta**2
sym.plot(s_y, (s_beta, -2,2), legend=True, title=f'${sym.latex(s_y)}$', y_label=None, x_label='x')
```
