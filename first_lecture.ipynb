```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(-3, 3, 51)
X, Y = np.meshgrid(xx, xx)


f1 = 4
f2 = .8
p1 = .6
p2 = -.77


Z = np.sin(f1*X*np.cos(f2*Y+p2)+p1)
gx, gy = np.gradient(Z)
```

The arrows correspond to the gradient which point to positive values

```python
plt.contourf(xx, xx, Z, 40, cmap='Spectral_r')
plt.quiver(xx, xx, gx, gy)
```
